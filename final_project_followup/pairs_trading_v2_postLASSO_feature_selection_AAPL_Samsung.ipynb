```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np

import statsmodels
import statsmodels.api as sm

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_prices_all = pd.read_csv('top100cap_stock_prices.csv',parse_dates=['Date'],index_col='Date')
dates_list = ['2019-01-01', '2019-04-01', '2019-07-01', '2019-10-01', 
              '2020-01-01', '2020-04-01', '2020-07-01', '2020-10-01',
              '2021-01-01', '2021-04-01', '2021-07-01', '2021-10-01',
              '2022-01-01', '2022-04-01', '2022-07-01', '2022-10-01', '2023-01-01']
tickers = ['005930.KS', 'AAPL'] #x,y
df_prices = df_prices_all[tickers]
```

We first plot the pairs trading strategy. We train our model using 4 year trailing data (blue), to predict spread in 6-month OoS price spread (orange), and extend that further in case there are existing positions at the end of the 6-month OoS period that require closing (green). 

```python
#each list corresponds to data for each OoS period
beta_list = [] #OLS betas for use in analysis later on (for each OoS period)
train_spread_std_list = [] #standard deviation of spread from training data
train_normalized_spread_ceil_list = [] #to determine what ratio of portfolio to invest and what to hold in cash-heavy assets later on

for i in range(len(dates_list)-9): #one year in-sample data, 3 months out-of-sample
    
    train_data = df_prices[dates_list[i]:dates_list[i+8]] #in-sample data
    test_data = df_prices[dates_list[i+8]:dates_list[i+9]] #out-of-sample data
    
    x = sm.add_constant(train_data[tickers[0]], prepend=False)
    ols = sm.OLS(train_data[tickers[1]], x).fit()
    print(ols.summary2())
    beta = ols.params
    beta_list.append(beta) #store values for later use

    train_spread = pd.DataFrame(train_data[tickers[1]] - beta[0] * train_data[tickers[0]] - beta[1])
    train_spread.columns = ['in-sample']
    
    #store values for later use
    train_spread_mean = train_spread['in-sample'].mean() #=0
    train_spread_std = train_spread['in-sample'].std()
    train_spread_std_list.append(train_spread_std)
    
    train_spread_max_normed = train_spread['in-sample'].max() / train_spread['in-sample'].std()
    train_spread_min_normed = train_spread['in-sample'].min() / train_spread['in-sample'].std()
    abs_norm_spread_max = 1 + max(np.abs(train_spread_max_normed), np.abs(train_spread_min_normed))
    train_normalized_spread_ceil_list.append(np.ceil(abs_norm_spread_max))

    test_spread = pd.DataFrame(test_data[tickers[1]] - beta[0] * test_data[tickers[0]] - beta[1])

    plt.figure(figsize=(10,6))
    plt.grid(True)
    plt.grid(which='minor', linestyle=':', linewidth='0.5')
    plt.plot(train_spread,linewidth=2)
    plt.plot(test_spread,label='out of sample',linewidth=2)
    
    test_data2 = pd.DataFrame() #out-of-sample data2 extends 3-month oos horizon to exit any pending position
    if i+9 < len(dates_list)-1:
        test_data2 = df_prices[dates_list[i+9]:dates_list[len(dates_list)-1]]
        test_spread2 = pd.DataFrame(test_data2[tickers[1]] - beta[0] * test_data2[tickers[0]] - beta[1])
        plt.plot(test_spread2,label='out of sample2',linewidth=2)
        
    plt.axhline(train_spread_mean, ls = '--', color='r',linewidth=3)
    plt.axhline(train_spread_mean + train_spread_std, ls = '--', color='g',linewidth=2)
    plt.axhline(train_spread_mean - train_spread_std, ls = '--', color='g',linewidth=2)
    plt.ylabel('Spread')
    plt.title('AAPL-Samsung Spread')
    plt.xlabel('Date since ' + str(train_data.index[0])[:10])
    plt.show()
    plt.close()
```

Observe how all of the t-stats are extremely statistically signficant for our analysis.

Observe how we generate a ceiling on the max/min of the spread of training sample for each time frame, which we will use as a metric for when to book exit a position in the out-of-sample period. For the out-of-sample period, when the normalized spread is above the value dictated by the maximum spread in-training data observed among t training samples, then we close the position and book a loss. This can be expressed mathematically as follows:  
  

  
The green data shows volatility beyond our training horizon and we often observe that spread is mean-reverting in the long run. By taking the max training-spread observed upto time t, we hope to acquire some long-term knowledge of this volatility over more horizons increase our confidence in how long to hold a position with the expectation of mean reversion.

```python
df_oos = df_prices.copy().loc['2021-01-01':]
df_spread = pd.DataFrame(index=df_oos.index)
df_spread_norm = pd.DataFrame(index=df_oos.index)
# tickers
# beta_list
# train_spread_std_list
# train_normalized_spread_ceil_list
dates_list_oos = ['2021-01-01', '2021-04-01', '2021-07-01', '2021-10-01',
                  '2022-01-01', '2022-04-01', '2022-07-01', '2022-10-01', '2023-01-01']

for i in range(len(dates_list_oos)-1):
    temp_df = df_oos.loc[dates_list_oos[i]:dates_list_oos[i+1]]
    beta = beta_list[i]
    temp_df[dates_list_oos[i]] = temp_df[tickers[1]] - beta[0] * temp_df[tickers[0]] - beta[1]
    df_spread[dates_list_oos[i]] = temp_df[dates_list_oos[i]]

for i in range(len(dates_list_oos)-1):
    df_spread_norm[dates_list_oos[i]] = df_spread[dates_list_oos[i]] / train_spread_std_list[i]
```

The data above explains the normalized spread for the 3-months out-of-sample period. Notice how the absolute values of normalized spread here are within 4 standard deviations. In our strategy, we go L/S when the absolute value of OOS spread exceeds one standard deviation, and revert our positions when spread goes back to 0, i.e. the sign of the normalized spread crosses from - to + or from + to -.  
  
In the event that we go L/S near the end of the 3-month OoS period, we extend our original 3-month model further if possible, until the position can be returned. In our model, we don't go L/S on a new position in the next OoS model period until the past one has been reverted, to avoid taking on excess leverage. 
  
In this analysis, we avoid dealing with modelling margin calls. Next, we investigate the absolute value of our normalized spread when extending our models past the 3-month OoS period to get an idea for how far the model can deviate should we fail to update our models every 3 months.

```python
df_oos = df_prices.copy().loc['2021-01-01':]
df_spread = pd.DataFrame(index=df_oos.index)
df_spread_norm = pd.DataFrame(index=df_oos.index)
# tickers
# beta_list
# train_spread_std_list
# train_normalized_spread_ceil_list

for i in range(len(dates_list_oos)-1):
    temp_df = df_oos.loc[dates_list_oos[i]:]
    beta = beta_list[i]
    temp_df[dates_list_oos[i]] = temp_df[tickers[1]] - beta[0] * temp_df[tickers[0]] - beta[1]
    df_spread[dates_list_oos[i]] = temp_df[dates_list_oos[i]]

for i in range(len(dates_list_oos)-1):
    df_spread_norm[dates_list_oos[i]] = df_spread[dates_list_oos[i]] / train_spread_std_list[i]
```

We observe that models predicting out-of-sample periods seem to work upto around the next one year - as evidenced by the fact that the rightmost five columns have a normalized spread of mostly under four standard deviations from min/max and the normalized spread model we created also seems to cover this with upto five standard deviations. We observe a smaller range of spreads for more recent models, and that's good for my strategy because it means the chances of margin calls are far smaller if the modelled spread is not too crazy.  
  
The leftmost three columns also show us how dangerous not updating the parameters of the model regularly is, as the oldest model sees a maximum of over seven standard deviations of normalized spread for max/min during the entire out-of-sample period. Thus, if a position is still open, past the three-month out-of-sample period, our goal should be to close it at the earliest possible opportunity.  
  
Thus, while in the three-month out-of-sample period, we said our strategy should sell its position when the spread returns to 0, if the trend is open past the three-month period I relax this condition to the absolute value of the normalized spread returning within 0.25, so as close out a position sooner and move on with the new updated model. Finally, for this simple program, I will not take new positions until having closed out old positions, so as not to introduce excess leverage when investing.

My trading strategy can be summarized as follows: 
- When the normalized spread > +1 and <max norm spread ceiling from training history, short \\$1 worth of AAPL long \\$1 worth of Samsung, revert position when normalized spread first becomes a negative number (i.e. reaches 0).
- When the normalized spread < -1 and >max norm spread ceiling from training history, long \\$1 worth of AAPL short \\$1 worth of Samsung, revert position when normalized spread first becomes a positive number (i.e. reaches 0).
- If position is not closed at the end of three-month out-of-sample period (solid orange lines in above diagrams), maintain position with the old model (solid green lines in above diagrams) until absolute value of normalized spread is within 0.25 with the same old model.
- Do not take a new position in an updated model until past position has been exited.
- If on first day of OoS we have absolute value of normalized spread greater than 1 and we don't have an existing position, enter market position immediately.  
  
To compare the previous day with the current, I lag my data by one day and compare previous to current. I act on the conditions mentioned above the next trading day, so lag once more afterwards.

```python
for i in range(len(dates_list_oos)-1):
    df_spread_norm['lag '+dates_list_oos[i]] = df_spread_norm[dates_list_oos[i]].shift()
df_spread_norm
```

Next, we collect a list of a tuple/dictionary of dates+values, which correspond to the date being the date we take action on and values (-1, 0, 1), which corresponds to whether to (long AAPL short Samsung, revert, short AAPL long Samsung) respectively. Some considerations: -1 for a date must be followed by a 0 the next date, else we cannot take a new position according to how we defined our strategy.

```python
# tickers
# beta_list
# train_spread_std_list
# train_normalized_spread_ceil_list

date_action_list = []
for i in range(len(dates_list_oos)-1):
    
    temp_df = df_spread_norm[['lag '+dates_list_oos[i], dates_list_oos[i]]]
    temp_df = temp_df.dropna(axis=0)
    
    should_i_trade = True
    if date_action_list: #if date_action_list is not empty
        
        #check if most recent action already exceeds the start date of this OoS sample, 
        #if so we need to start at later time to avoid excess leverage
        most_recent_action_date = date_action_list[len(date_action_list)-1][0]
        most_recent_action_type = date_action_list[len(date_action_list)-1][1]
        
        if most_recent_action_type == 0:
            #most recent reversion occurs after start of OoS date range, so take no new action till that date
            if most_recent_action_date > temp_df.index[0]: 
                while most_recent_action_date != temp_df.index[0]:
                    temp_df = temp_df.drop(str(temp_df.index[0])[:10],axis=0)
        
        #that means that we have taken a previous position and haven't closed it due to lack of signal, 
        #so do not make any new trades
        if most_recent_action_type != 0:
            should_i_trade = False
    
    reverted_status = True
    while len(temp_df!=0) and should_i_trade: #we will examine each temp_df row-by-row
        
        if reverted_status == True: #if we are not trading already
            
            #explore adding new position only if OoS sample is within 3 months, otherwise we move on to next 3-month OLS model
            if temp_df.index[0] < pd.Timestamp(dates_list_oos[i+1]):
            
                first_spread = temp_df.iat[(0,1)]
                #if absolute value >1 stdev for OoS spread and less than max observed norm spread from training, 
                #record date and entry position
                if np.abs(first_spread) > 1 and \
                np.abs(first_spread) < max(train_normalized_spread_ceil_list[:i+1]): 
                    date_action_list.append( (temp_df.index[0], np.sign(first_spread)) )
                    reverted_status = False
            
        else: #we are in a pairs-trade position, and need to check if it's time to close position
            
            compare_row = temp_df.iloc[0].tolist()

            if temp_df.index[0] < pd.Timestamp(dates_list_oos[i+1]): #check if position is open before next OoS period
            
                #if exceed max(train_normalized_spread_ceil_list) upto time t/i standard deviations on normalized spread, 
                #exit (book major loss)
                if np.abs(compare_row[0]) > max(train_normalized_spread_ceil_list[:i+1]): 
                    date_action_list.append( (temp_df.index[0], 0) )
                    reverted_status = True
                    temp_df = temp_df.drop(str(temp_df.index[0])[:10],axis=0)
                    continue #i.e. skip to start of while loop

                #this only happens if the prev/current spread sign changes, i.e. -/+ or +/-
                if sum(np.sign(compare_row)) == 0:
                    date_action_list.append( (temp_df.index[0], 0) )
                    reverted_status = True
            
            else: #position is open after next OoS period has started
                
                #exit position to book loss if normalized spread exceeds max(train_normalized_spread_ceil_list) upto time t/i
                if np.abs(compare_row[0]) > max(train_normalized_spread_ceil_list[:i+1]): 
                    date_action_list.append( (temp_df.index[0], 0) )
                    reverted_status = True
                    temp_df = temp_df.drop(str(temp_df.index[0])[:10],axis=0)
                    continue #i.e. skip to start of while loop
                
                #exit position if absolute normalized spread is within 0.25
                if np.abs(compare_row[0]) < 0.25: 
                    date_action_list.append( (temp_df.index[0], 0) )
                    reverted_status = True
            
        temp_df = temp_df.drop(str(temp_df.index[0])[:10],axis=0)
```

We see that based on our trade strategy, we execute 3 trades over the past nearly 24 months of OoS analysis, having exited from 3 trade positions. Finally, now we look at the price of Apple and Microsoft on the date after those dates, using that to determine how much money we made through basic statistical arbitrage/our pairs-trading strategy. We use the date afer those dates, because the date we have is the adjusted signal generated from the adjusted closing price that day, and we act on the signal the next day, hence the date after those dates.  
  
When the normalized spread > +1, short \\$1 worth of  AAPL long \\$1 worth of Samsung, revert position when normalized spread first becomes a negative number.  
When the normalized spread < -1, long \\$1 worth of AAPL short \\$1 worth of Samsung, revert position when normalized spread first becomes a positive number.
- action_type = 1 -> short AAPL, long Samsung (in equal quantities of cash)
- action_type = -1 -> long AAPL, short Samsung (in equal quantities of cash)
- action_type = 0 -> return ownership of #shares of both stocks to 0

```python
action_dates, action_type = zip(*date_action_list)
action_dates = [str(i)[:10] for i in action_dates]
```

```python
df_prices_lagged_up = pd.DataFrame(index=df_prices.index)
df_prices_lagged_up = df_prices.shift(-1) #use the next day's price after observing signal on current day

df_execution_days = df_prices_lagged_up.loc[action_dates]
df_execution_days['action type'] = action_type

#returns for $100 traded equally doing L/S between Samsung and AAPL
df_execution_days['Samsung ownership'] = (df_execution_days['action type']*100) / df_execution_days['005930.KS']
df_execution_days['AAPL ownership'] = (df_execution_days['action type']*-100) / df_execution_days['AAPL']
df_execution_days['arbitrage returns'] = df_execution_days['AAPL']*df_execution_days['AAPL ownership'].shift() + \
                                            df_execution_days['005930.KS']*df_execution_days['Samsung ownership'].shift()
df_execution_days
```
